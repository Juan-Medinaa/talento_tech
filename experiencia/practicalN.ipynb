```python
import pandas as pd

basepath = "/content/IRIS.csv"
df = pd.read_csv(basepath)
df.head()
```

```python
#crear graficos de barra para mostrar el promedio de longitud del sepalo por especie
sns.barplot(data=df, x='species', y='sepal_length')
plt.title('Promedio de longitud del sepalo por especie')
plt.xlabel('Especie')
plt.ylabel('Promedio de Longitud del Sépalo')
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal_width', y='petal_width', hue='species')
plt.title('Diagrama de Dispersión: Ancho del Sépalo vs Ancho del Pétalo')
plt.xlabel('Ancho del Sépalo')
plt.ylabel('Ancho del Pétalo')
plt.legend(title='Especie', loc='upper right', fontsize='small')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
#from sklearn.feature_selection
from sklearn.decomposition import PCA

data = load_iris()
X = data.data
#Y = data.target

pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X)

print(X_reduced[:5])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming `data` (from load_iris) and `X_reduced` are available from previous cells.
# If `data` is not explicitly available, we can reload it.
from sklearn.datasets import load_iris
data = load_iris()

df_pca = pd.DataFrame(data=X_reduced, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Species'] = data.target
df_pca['Species Name'] = df_pca['Species'].apply(lambda x: data.target_names[x])

plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Species Name',
    data=df_pca,
    palette='viridis',
    s=100,
    alpha=0.8
)
plt.title('2 Component PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
dfa_corr = df_pca.drop(columns=['Species', 'Species Name'])
corr_matrix = dfa_corr.corr()
print(corr_matrix)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
